```python
import pandas as pd
import sqlite3
```

```python

def ejecutar_consulta(query, db_path='../data/data.db'):
    conn = sqlite3.connect(db_path)
    df = pd.read_sql_query(query, conn)
    conn.close()
    return df
```

```python
df = ejecutar_consulta("SELECT * FROM producto")
```

```python
df.head(10)
```

```python
query = """
SELECT customer_id, product_id, periodo, tn
FROM sellin
"""
ventas_df = ejecutar_consulta(query)

pivot_df = ventas_df.pivot_table(
    index=['customer_id', 'product_id'],
    columns='periodo',
    values='tn'
).reset_index()

pivot_df.columns.name = None
pivot_df = pivot_df.rename_axis(None, axis=1)

pivot_df.head()
```
